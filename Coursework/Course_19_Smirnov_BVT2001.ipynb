```python
from itertools import count

def lastmax(sequence):
    return max(zip(sequence, count()))

n = int(input("–ö–æ–ª-–≤–æ –≥–æ—Ä–æ–¥–æ–≤: "))
an = []
print(f"–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ —Å–µ–ª—Ñ–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ {n} –≥–æ—Ä–æ–¥–æ–≤:")

selfies = input().split(" ")

for i in selfies:
    an.append(i)
        
d1 = {}

for i in range(1,len(an)+1):
    d1[i] = int(an[i-1])

max_selfie = max(d1.values())

c = lastmax(d1.values())[1]+1
res = max_selfie*n - (n-c)
print(res)
                    

```

```python
def bobr(n, trees):
    a,b = {}, {}
    for i, j in enumerate(trees):
        if j not in a:
            a[j] = i
        else:
            b[j] = i
    p = []
    for j in b:
        p.append((a[j], b[j]-1))
    s = []
    for j in trees:
        if j > 0:
            s.append(j)
        else:
            s.append(0)
    u = v = 2 * trees[p[0][0]]-1
    for i in range(n-1):
        s[i+1] += s[i]
    for i,j in p:
        u = 2*trees[i] + s[j] - s[i]
        if u>v:
            a = i
            b = j
            v = u
    s = list(range(1, a + 1)) + [i for i, j in enumerate(trees[a + 1: b + 1], a + 2) if j < 0] + list(range(b + 3, n + 1))
    res = str(v) + ' ' + str(len(s)) + '\n' + ' '.join(map(str, s))
    return res
    
n, trees = int(input()), list(map(int, input().split()))
print(bobr(n, trees))

         
                
```

```python
def interpreter():
    n, q = map(int, input().split(" "))
    program = input()
    result = []
    for k in range(q):
        l, r = map(int, input().split(" "))
        temp_program = list(program[l - 1: r])
        cp = 0
        dp = 1
        temp_result = [0 for i in range(10)]
        flag = False
        while 0 <= cp < len(temp_program):
            current_symbol = temp_program[cp]
            if current_symbol.isdigit():
                flag = False
                if current_symbol == "0":
                    temp_result[0] += 1
                    temp_program.pop(cp)
                    if dp == 1:
                        cp -= 1
                else:
                    temp_result[int(temp_program[cp])] += 1
                    temp_program[cp] = str(int(current_symbol) - 1)
            else:
                if flag:
                    temp_program.pop(cp - dp)
                    if dp == 1:
                        cp -= dp
                if current_symbol == ">":
                    dp = 1
                else:
                    dp = -1
                flag = True
            cp += dp
        result.append(temp_result)
 
    return result
 
result = interpreter()
for line in result:
    print(" ".join([str(el) for el in line]) + " ")
```

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ª–µ–¥—É–µ—Ç —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–≤–µ–∑–µ–Ω–Ω—ã—Ö –±—É–∫–µ—Ç–æ–≤.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ª–µ–¥—É–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª a1,‚Äâa2,‚Äâ...,‚Äâan (1‚Äâ‚â§‚Äâai‚Äâ‚â§‚Äâ106) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–≤–µ—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–º –∏–∑ –ø—Ä–∏–≤–µ–∑–µ–Ω–Ω—ã—Ö –±—É–∫–µ—Ç–æ–≤.

```python
def max_bouquette(n):
    odd =0
    even = 0
    bouq = input().split(" ")[:n]
    for i in bouq:
        if int(i) % 2 ==0:
            even += 1
        else:
            odd += 1
    res = min(even, odd) + max(0, (odd - even) // 3)
    return res
print(max_bouquette(int(input())))
```

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

```python
from itertools import combinations

def queue_num(n,x, arr, ceps, ones, s):
    for i in range(1, n+1):
        if i not in s and i not in arr:
            cep = [i]
            while arr[i]:
                cep.append(arr[i])
                i = arr[i]
            for i in cep:
                s.add(i)
            if x in cep:
                right = cep.index(x)
                left = len(cep) - right - 1
            else:
                if len(cep) == 1:
                    ones+=1
                else:
                    ceps.append(len(cep))
    sums = set(ceps)
    for i in range(2, len(ceps)+1):
        for comb in combinations(ceps,i):
            sums.add(sum(comb))
    sums.add(0)
    
    pos_set = set()
    for s1 in sums:
        for i in range(ones+1):
            pos_set.add(left + s1 + 1 + i)
    return sorted(pos_set)
                    
n,x = map(int, input().split(" "))
arr = [-10000] + list(map(int, input().split(" ")))
ceps = []
ones = 0
s = set()

for i in queue_num(n,x,arr,ceps,ones,s):
    print(i)
```

```python
def crossroads(n):
    g=[[] for i in range(n)]
    for _ in range(n-1):
        x,y=map(int,input().split(" "))
        x-=1
        y-=1
        g[x].append(y)
        g[y].append(x)
    subtree_size=[0]*n
    stack=[[-1,0,0]]
    ans=[]
    while stack:
        par,ver,state=stack.pop()
        if state==0:
            stack.append([par,ver,1])
            for to in g[ver]:
                if to!=par:
                    stack.append([ver,to,0])
        else:
            if len(g[ver])==1:
                subtree_size[ver]=1
            else:
                cnt=0
                tmp=[]
                for to in g[ver]:
                    if to!=par:
                        cnt+=subtree_size[to]
                        tmp.append(subtree_size[to])
                tmp.append(n-cnt-1)
                local=[0]*(n+1)
                local[0]=1
                for x in tmp:
                    for i in range(n-x,-1,-1):
                            if local[i]==1:
                                local[i+x]=1
                                if x+i!=0 and n-1-(x+i)!=0:
                                    ans.append(x+i)
                subtree_size[ver]=cnt+1
    ans=sorted(list(set(ans)))
    return ans

n=int(input())

ans = crossroads(n)
print(len(ans))
for i in ans:
    print(i,n-1-i)
```
