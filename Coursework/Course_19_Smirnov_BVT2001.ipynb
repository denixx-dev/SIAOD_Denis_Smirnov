```python
from itertools import count

def lastmax(sequence):
    return max(zip(sequence, count()))

n = int(input("–ö–æ–ª-–≤–æ –≥–æ—Ä–æ–¥–æ–≤: "))
an = []
print(f"–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ —Å–µ–ª—Ñ–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ {n} –≥–æ—Ä–æ–¥–æ–≤:")

selfies = input().split(" ")

for i in selfies:
    an.append(i)

    
d1 = {}


for i in range(1,len(an)+1):
    d1[i] = int(an[i-1])


max_selfie = max(d1.values())


c = lastmax(d1.values())[1]+1
res = max_selfie*n - (n-c)
print(res)
                    

```

```python
def bobr(n, trees):
    a,b = {}, {}
    for i, j in enumerate(trees):
        if j not in a:
            a[j] = i
        else:
            b[j] = i
    p = []
    for j in b:
        p.append((a[j], b[j]-1))
    s = []
    for j in trees:
        if j > 0:
            s.append(j)
        else:
            s.append(0)
    u = v = 2 * trees[p[0][0]]-1
    for i in range(n-1):
        s[i+1] += s[i]
    for i,j in p:
        u = 2*trees[i] + s[j] - s[i]
        if u>v:
            a = i
            b = j
            v = u
    s = list(range(1, a + 1)) + [i for i, j in enumerate(trees[a + 1: b + 1], a + 2) if j < 0] + list(range(b + 3, n + 1))
    res = str(v) + ' ' + str(len(s)) + '\n' + ' '.join(map(str, s))
    return res
    
n, trees = int(input()), list(map(int, input().split()))
print(bobr(n, trees))

         
                
```

```python
def smart_bobr(n,m, arr):
    fib = [0 for i in range(101)]
    fib[0] = fib[1] = 1
    for i in range(2,101):
        fib[i]=fib[i-1]+fib[i-2]
    for i in range(m):
        ops=str(input())
        if ops[0]=="1":
            qs=ops.split(" ");
            xi=int(qs[1])-1
            vi=int(qs[2])
            arr[xi]=vi
        elif ops[0]=="2":
            qs=ops.split(" ")
            li=int(qs[1])-1
            ri=int(qs[2])-1
            ans=0
            for x in range(0,ri-li+1):
                ans=ans+fib[x]*arr[li+x]
            print (ans%1000000000)
        else:
            qs=ops.split(" ")
            li=int(qs[1])-1
            ri=int(qs[2])-1
            d=int(qs[3])
            for i in range(li,ri+1):
                arr[i]+=d

n,m = map(int, input().split(" "))
arr = input().split(" ")[:n]
for i in range(len(arr)):
    arr[i] = int(arr[i])
smart_bobr(n,m,arr)
```
